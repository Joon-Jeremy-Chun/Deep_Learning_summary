```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [[[i +j] for i in range(5)] for j in range(100)]

X[:3],X[-3:]
```

```python
y = [(i +5) for i in range(100)]

y[:3],y[-3:]
```

```python
X = np.array(X, dtype = float)
y = np.array(y, dtype = float)
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
  random_state = 2045)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from tensorflow.keras import models, layers

model_1 = models.Sequential(name = 'SimpleRNN_1')
model_1.add(layers.SimpleRNN(3,
                             input_shape = (5,1),
                             return_sequences = False))
model_1.add(layers.Dense(1))

model_1.summary()
```

```python
model_1.compile(loss = 'mse',
                optimizer = 'adam',
                metrics = ['accuracy'])
```

```python
Hist_1 = model_1.fit(X_train, y_train,
                     epochs = 100,
                     batch_size = 8,
                     validation_data = (X_test,y_test))
```

```python
y_hat = model_1.predict(X_test)
```

```python
plt.plot(Hist_1.history['loss'])
plt.show()
```

```python
plt.scatter(range(20), y_test, c = 'g')
plt.scatter(range(20), y_hat, c = 'r')
plt.show()
```

```python
Hist_2 = model_1.fit(X_train, y_train,
                     epochs = 500,
                     batch_size = 8,
                     validation_data = (X_test,y_test))
```

```python
y_hat = model_1.predict(X_test)
```

```python
plt.plot(Hist_1.history['loss'])
plt.show()
```

```python
plt.scatter(range(20), y_test, c = 'g')
plt.scatter(range(20), y_hat, c = 'r')
plt.show()
```

```python
X = [[[(i + j)] for i in range(5)] for j in range(100)]
y = [(i + 5) for i in range(100)]
X = (X - np.min(X)) / (np.max(X) - np.min(X))
y = (y - np.min(y)) / (np.max(y) - np.min(y))
```

```python
X = np.array(X, dtype = float)
y = np.array(y, dtype = float)
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
  random_state = 2045)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
model_2 = models.Sequential(name = 'SimpleRNN_2')
model_2.add(layers.SimpleRNN(3,
                             input_shape = (None, 1), 
                             return_sequences = False))
model_2.add(layers.Dense(1))
model_2.summary()
```

```python
model_2.compile(loss = 'mse', 
                optimizer = 'adam', 
                metrics = ['accuracy'])
```

```python
Hist_2 = model_2.fit(X_train, y_train, 
                     epochs = 100,
                     batch_size = 8,
                     validation_data = (X_test, y_test))
```

```python
y_hat = model_2.predict(X_test)
```

```python
plt.plot(Hist_2.history['loss'])
plt.show()
```

```python
plt.scatter(range(20), y_test, c = 'g')
plt.scatter(range(20), y_hat, c = 'r')
plt.show()
```

# III. 레이어를 한단계더

```python
model_3 = models.Sequential(name = 'Stackd_RNN')
model_3.add(layers.SimpleRNN(3,
                             input_shape = (None, 1), 
                             return_sequences = True))
model_3.add(layers.SimpleRNN(3,
                             input_shape = (None, 1),
                             return_sequences = False))
model_3.add(layers.Dense(1))
model_3.summary()
```

```python
model_3.compile(loss = 'mse', 
                optimizer = 'adam', 
                metrics = ['accuracy'])
```

```python
Hist_3 = model_3.fit(X_train, y_train, 
                     epochs = 100,
                     batch_size = 8,
                     validation_data = (X_test, y_test))
```

```python
y_hat = model_3.predict(X_test)
```

```python
plt.plot(Hist_3.history['loss'])
plt.show()
```

```python
plt.scatter(range(20), y_test, c = 'g')
plt.scatter(range(20), y_hat, c = 'r')
plt.show()
```
